Search a 2d Matrix

Binary search

You are given an m x n integer matrix matrix with the following two properties:

- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer target, return true if target is in matrix or false otherwise.

You must write a solution in O(log(m \* n)) time complexity.

```python
def searchMatrix(matrix: list[list[int]], target: int) -> bool:
    height = len(matrix)  # amount of rows
    width = len(matrix[0])  # amount of columns

    # row, column
    left = [0, 0]
    right = [height - 1, width - 1]

    # tuples/list are compared elementwise
    while left <= right:
        # special case when it converges to 0,0
        if left == right:
            return matrix[left[0]][left[1]] == target

        # convert coordinates to how many elements are left to check
        total = (right[0] - left[0] + 1) * width
        if total != height * width:
            total += right[1] - left[1] + 1
        total_center = total // 2

        # problematic when width is 1
        # find the coordinates of centeral element
        rows_to_check = total_center // width
        # cols_to_check = (total_center % width) - 1
        # cols_to_check = rows_to_check * width - total_center
        cols_to_check = total_center % width

        # center = [left[0] + rows_to_check - 1, left[1] + cols_to_check]
        center = [left[0] + rows_to_check, left[1] + cols_to_check - 1]
        if width == 1:
            center = [left[0] + rows_to_check - 1, left[1] + cols_to_check]

        val = matrix[center[0]][center[1]]
        if val < target:
            center[1] += 1
            if center[1] >= width:
                center[0] += 1
                center[1] = 0

            left = center
        elif val > target:
            center[1] -= 1
            if center[1] < 0:
                center[0] -= 1
                center[1] = width - 1

            right = center
        else:
            return True

    return False


"""
    left = 0
    right = len(nums) - 1

    while left <= right:
        center = (left + right) // 2
        if nums[center] < target:
            left = center + 1
        elif nums[center] > target:
            right = center - 1
        else:
            return center

    return -1
"""
assert searchMatrix([[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], 60) == True
assert searchMatrix([[1, 3]], 2) == False
assert searchMatrix([[1, 3]], 1) == True
assert searchMatrix([[1, 3]], 3) == True
assert searchMatrix([[1], [3]], 2) == False
assert searchMatrix([[1], [3]], 1) == True
assert searchMatrix([[1], [3]], 3) == True
assert searchMatrix([[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], 1) == True
assert searchMatrix([[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], 13) == False
assert searchMatrix([[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], 3) == True
```
